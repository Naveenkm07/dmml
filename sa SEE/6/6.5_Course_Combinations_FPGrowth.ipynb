### a. Prepare Transaction Dataset

```python
dataset = [
    ['Math', 'Physics', 'Computer Science'],
    ['Math', 'Computer Science', 'Statistics'],
    ['Physics', 'Chemistry', 'Math'],
    ['Computer Science', 'Statistics', 'Physics'],
    ['Math', 'Computer Science']
]
print('BEFORE ENCODING (Raw Transactions):')
print(dataset)
```

### Convert to Transactional One-Hot Encoded DataFrame

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

te = TransactionEncoder()
df = pd.DataFrame(te.fit_transform(dataset), columns=te.columns_)
print('AFTER ONE-HOT ENCODING:')
display(df.head())
```

### b. Mine Frequent Itemsets using FP-Growth

```python
from mlxtend.frequent_patterns import fpgrowth

freq_itemset = fpgrowth(df, min_support=0.4, use_colnames=True)
display(freq_itemset)
```

### c. Generate Association Rules

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(freq_itemset, metric='confidence', min_threshold=0.5)

display(rules)
```

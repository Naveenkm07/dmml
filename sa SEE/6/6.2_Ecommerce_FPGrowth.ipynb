### a. Prepare Transaction Dataset

```python
dataset = [
    ['laptop', 'mouse', 'keyboard'],
    ['laptop', 'mouse'],
    ['monitor', 'mouse', 'keyboard'],
    ['laptop', 'monitor', 'keyboard'],
    ['mouse', 'keyboard']
]
print('BEFORE ENCODING (Raw Transactions):')
print(dataset)
```

### Convert to Transactional One-Hot Encoded DataFrame

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

te = TransactionEncoder()
te_data = te.fit_transform(dataset)
df = pd.DataFrame(te_data, columns=te.columns_)
print('AFTER ONE-HOT ENCODING:')
display(df.head())
```

### b. Mine Frequent Itemsets using FP-Growth

```python
from mlxtend.frequent_patterns import fpgrowth

freq_itemset = fpgrowth(df, min_support=0.4, use_colnames=True)
display(freq_itemset)
```

### c. Generate Association Rules

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(freq_itemset, metric='confidence', min_threshold=0.5)
display(rules)
```

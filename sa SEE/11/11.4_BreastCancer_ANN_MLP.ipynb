```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, recall_score
import seaborn as sns
import matplotlib.pyplot as plt
```

### a. Standardize features and encode labels (Malignant=1, Benign=0)

```python
data = load_breast_cancer()
X = data.data
y = 1 - data.target 

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### b & c. Build ANN (two hidden layers), train with stratified split

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, stratify=y, random_state=42)

mlp = MLPClassifier(hidden_layer_sizes=(10, 5), max_iter=500, random_state=42)
mlp.fit(X_train, y_train)

y_pred = mlp.predict(X_test)

sensitivity = recall_score(y_test, y_pred)
print(f"Sensitivity (Recall): {sensitivity:.2f}")

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cmap='Reds', fmt='d')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```

### d. Perform k-fold cross-validation

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_scores = cross_val_score(mlp, X_scaled, y, cv=skf, scoring='accuracy')

print(f"K-Fold CV Accuracies: {cv_scores}")
print(f"Average Performance: {np.mean(cv_scores):.4f}")
```

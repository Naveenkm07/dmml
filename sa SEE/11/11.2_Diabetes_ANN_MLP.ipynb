```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.neural_network import MLPClassifier
```

### a. Handle missing values and scale features

```python
data = {
    'Pregnancies': [6, 1, 8, 1, 0, 5],
    'Glucose': [148, 85, 183, 89, np.nan, 116],
    'BloodPressure': [72, 66, 64, np.nan, 40, 74],
    'BMI': [33.6, 26.6, 23.3, 28.1, 43.1, 25.6],
    'Outcome': [1, 0, 1, 0, 1, 0]
}
df = pd.DataFrame(data)

df.fillna(df.mean(), inplace=True)

X = df.drop('Outcome', axis=1)
y = df['Outcome']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

### b & c. Implement ANN with two hidden layers, train with early stopping

```python
mlp = MLPClassifier(hidden_layer_sizes=(12, 8), max_iter=200, early_stopping=True, random_state=42)
mlp.fit(X_train, y_train)
```

### d. Evaluate using accuracy, precision, recall, F1, ROC-AUC

```python
y_pred = mlp.predict(X_test)
y_prob = mlp.predict_proba(X_test)[:, 1]

print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
print(f"Precision: {precision_score(y_test, y_pred, zero_division=0):.2f}")
print(f"Recall: {recall_score(y_test, y_pred, zero_division=0):.2f}")
print(f"F1-Score: {f1_score(y_test, y_pred, zero_division=0):.2f}")
print(f"ROC-AUC: {roc_auc_score(y_test, y_prob):.2f}")
```

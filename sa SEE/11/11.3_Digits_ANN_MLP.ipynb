```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
```

### a. Normalize pixel values and one-hot encode

```python
digits = load_digits()
X = digits.data
y = digits.target

scaler = MinMaxScaler()
X_norm = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.3, random_state=42)
```

### b & c. Implement neural network with one hidden layer, mini-batch GD

```python
mlp = MLPClassifier(hidden_layer_sizes=(50,), activation='relu', solver='sgd',
                    batch_size=32, learning_rate_init=0.01, max_iter=300, random_state=42)

mlp.fit(X_train, y_train)
```

### d. Report accuracy and show misclassified images

```python
train_acc = accuracy_score(y_train, mlp.predict(X_train))
test_pred = mlp.predict(X_test)
test_acc = accuracy_score(y_test, test_pred)

print(f"Training Accuracy: {train_acc * 100:.2f}%")
print(f"Test Accuracy: {test_acc * 100:.2f}%")

misclassified_idx = np.where(test_pred != y_test)[0]

plt.figure(figsize=(10, 4))
for i, idx in enumerate(misclassified_idx[:4]):
    plt.subplot(1, 4, i + 1)
    plt.imshow(X_test[idx].reshape(8, 8), cmap='gray')
    plt.title(f"True: {y_test[idx]}\nPred: {test_pred[idx]}")
    plt.axis('off')
plt.tight_layout()
plt.show()
```

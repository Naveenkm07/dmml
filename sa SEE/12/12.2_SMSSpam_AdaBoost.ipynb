```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

### a. Feature Extraction

```python
data = {
    'Message': ['Win a free iPhone now', 'Hey, are we still meeting later?', 'URGENT: Your account is locked', 'Can you pick up milk?', 'Congratulations, you won 1000 dollars', 'See you tomorrow'],
    'Label': ['Spam', 'Ham', 'Spam', 'Ham', 'Spam', 'Ham']
}
df = pd.DataFrame(data)

X = df['Message'].apply(len).values.reshape(-1, 1)
y = df['Label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### b & c. Decision Tree and AdaBoost

```python
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test)

ada = AdaBoostClassifier(n_estimators=50, random_state=42)
ada.fit(X_train, y_train)
ada_pred = ada.predict(X_test)
```

### d & e. Compare results and identify best model

```python
print(f"Decision Tree Accuracy: {accuracy_score(y_test, dt_pred)*100:.2f}%")
print(f"AdaBoost Accuracy: {accuracy_score(y_test, ada_pred)*100:.2f}%")

fig, ax = plt.subplots(1, 2, figsize=(10, 4))
sns.heatmap(confusion_matrix(y_test, dt_pred), annot=True, cmap='Blues', ax=ax[0])
ax[0].set_title('Decision Tree CM')
sns.heatmap(confusion_matrix(y_test, ada_pred), annot=True, cmap='Greens', ax=ax[1])
ax[1].set_title('AdaBoost CM')
plt.show()

print("AdaBoost is typically the better model for text classification due to reducing bias.")
```

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

### a. Preprocessing and Normalization

```python
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)
```

### b & c. Implement Naive Bayes and Random Forest

```python
nb = GaussianNB()
nb.fit(X_train_sc, y_train)
nb_pred = nb.predict(X_test_sc)

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train_sc, y_train)
rf_pred = rf.predict(X_test_sc)
```

### d & e. Compare models and display table

```python
def get_metrics(y_true, y_pred):
    return [
        accuracy_score(y_true, y_pred),
        precision_score(y_true, y_pred),
        recall_score(y_true, y_pred),
        f1_score(y_true, y_pred)
    ]

results = pd.DataFrame([
    get_metrics(y_test, nb_pred),
    get_metrics(y_test, rf_pred)
], columns=['Accuracy', 'Precision', 'Recall', 'F1-Score'], index=['Naive Bayes', 'Random Forest'])

display(results)
print("\nRandom Forest generally performs better as an ensemble method.")
```

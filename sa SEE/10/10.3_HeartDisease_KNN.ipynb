```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

### a. Prepare Dataset & Split

```python
data = {
    'Age': [55, 30, 60, 45, 65, 35],
    'Cholesterol': [250, 180, 280, 210, 300, 190],
    'Blood Pressure': [140, 110, 150, 120, 160, 115],
    'Max Heart Rate': [120, 180, 110, 160, 105, 170],
    'Chest Pain Type': [3, 0, 3, 1, 3, 0],
    'Heart Disease': ['Yes', 'No', 'Yes', 'No', 'Yes', 'No']
}
df = pd.DataFrame(data)

le = LabelEncoder()
df['Heart Disease'] = le.fit_transform(df['Heart Disease'])

X = df.drop('Heart Disease', axis=1)
y = df['Heart Disease']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
df.head()
```

### b. Train k-NN Classifier & Predict New Sample

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_sc, y_train)

y_pred = knn.predict(X_test_sc)
print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.2f}%")

new_patient = pd.DataFrame([[50, 240, 130, 140, 2]], columns=X.columns)
new_patient_sc = sc.transform(new_patient)
pred_new = knn.predict(new_patient_sc)
print(f"\nNew Patient Prediction: {le.inverse_transform(pred_new)[0]}")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
```

### a. Prepare Dataset & Split

```python
data = {
    'Area (sq ft)': [1500, 800, 2500, 950, 3000, 1100],
    'Rooms': [3, 1, 4, 2, 5, 2],
    'Location Rating': [8, 5, 9, 6, 9, 4],
    'Price Category': ['Medium', 'Low', 'High', 'Low', 'High', 'Low']
}
df = pd.DataFrame(data)

le = LabelEncoder()
df['Price Category'] = le.fit_transform(df['Price Category'])

X = df.drop('Price Category', axis=1)
y = df['Price Category']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
df.head()
```

### b. Train k-NN Classifier & Predict New Sample

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_sc, y_train)

y_pred = knn.predict(X_test_sc)
print(f"Accuracy: {accuracy_score(y_test, y_pred)*100:.2f}%")

new_house = pd.DataFrame([[1800, 3, 7]], columns=X.columns)
new_house_sc = sc.transform(new_house)
pred_new = knn.predict(new_house_sc)
print(f"\nNew House Price Category: {le.inverse_transform(pred_new)[0]}")
```

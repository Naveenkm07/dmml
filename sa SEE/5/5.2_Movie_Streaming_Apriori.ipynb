### a. Prepare Transaction Dataset

```python
dataset = [
    ['Action', 'Sci-Fi', 'Thriller'],
    ['Action', 'Drama'],
    ['Sci-Fi', 'Thriller'],
    ['Action', 'Sci-Fi', 'Drama'],
    ['Action', 'Sci-Fi']
]
print(dataset)
```

### Convert to Transactional One-Hot Encoded DataFrame

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

te = TransactionEncoder()
df = pd.DataFrame(te.fit_transform(dataset), columns=te.columns_)
df.head()
```

### b. Run Apriori

```python
from mlxtend.frequent_patterns import apriori

freq_itemset = apriori(df, min_support=0.05, use_colnames=True)
freq_itemset['length'] = freq_itemset['itemsets'].apply(lambda x: len(x))
filtered_itemset = freq_itemset[(freq_itemset['length'] >= 2) & (freq_itemset['length'] <= 3)]
filtered_itemset
```

### c. Generate Association Rules

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(freq_itemset, metric='confidence', min_threshold=0.1)
genre_rules = rules[(rules['antecedents'].apply(lambda x: len(x) == 1)) & 
                    (rules['consequents'].apply(lambda x: len(x) == 1))]
genre_rules
```

### a. Prepare Transaction Dataset

```python
dataset = [
    ['Rain_No', 'Temp_High', 'Moisture_Low'],
    ['Rain_Light', 'Temp_Moderate', 'Moisture_High'],
    ['Rain_No', 'Temp_High', 'Moisture_Low'],
    ['Rain_Heavy', 'Temp_Low', 'Moisture_High'],
    ['Rain_No', 'Temp_Moderate', 'Moisture_Low']
]
print(dataset)
```

### Convert to Transactional One-Hot Encoded DataFrame

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

te = TransactionEncoder()
df = pd.DataFrame(te.fit_transform(dataset), columns=te.columns_)
df.head()
```

### b. Run Apriori

```python
from mlxtend.frequent_patterns import apriori

freq_itemset = apriori(df, min_support=0.2, use_colnames=True)
display(freq_itemset)
```

### c. Generate Association Rules

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(freq_itemset, metric='confidence', min_threshold=0.1)

display(rules)
```

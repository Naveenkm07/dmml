### a. Prepare Transaction Dataset

```python
dataset = [
    ['bread', 'milk', 'eggs', 'butter'],
    ['bread', 'milk', 'eggs'],
    ['milk', 'butter', 'cola'],
    ['bread', 'milk', 'butter', 'cola'],
    ['bread', 'milk']
]
print(dataset)
```

### Convert to Transactional One-Hot Encoded DataFrame

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

te = TransactionEncoder()
df = pd.DataFrame(te.fit_transform(dataset), columns=te.columns_)
df.head()
```

### b. Run Apriori

```python
from mlxtend.frequent_patterns import apriori

freq_itemset = apriori(df, min_support=0.02, use_colnames=True)
freq_itemset.head()
```

### c. Generate Association Rules

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(freq_itemset, metric='confidence', min_threshold=0.6)
rules.sort_values(by='confidence', ascending=False).head(5)
```

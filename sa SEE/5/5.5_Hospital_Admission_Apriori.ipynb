### a. Prepare Transaction Dataset

```python
dataset = [
    ['Fever', 'Cough', 'Fatigue'],
    ['Fever', 'Cough', 'Serious_Complication'],
    ['Fatigue', 'Headache'],
    ['Fever', 'Fatigue', 'Serious_Complication'],
    ['Cough', 'Fatigue']
]
print(dataset)
```

### Convert to Transactional One-Hot Encoded DataFrame

```python
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

te = TransactionEncoder()
df = pd.DataFrame(te.fit_transform(dataset), columns=te.columns_)
df.head()
```

### b. Run Apriori

```python
from mlxtend.frequent_patterns import apriori

freq_itemset = apriori(df, min_support=0.05, use_colnames=True)
freq_itemset
```

### c. Generate Association Rules

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(freq_itemset, metric='confidence', min_threshold=0.6)
complication_rules = rules[rules['consequents'].apply(lambda x: 'Serious_Complication' in x)]
complication_rules
```

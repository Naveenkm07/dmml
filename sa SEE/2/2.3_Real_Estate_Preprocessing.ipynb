```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
```

### a. Prepare Dataset (Mock Data)

```python
data = {
    'area': [1200, 1500, 1000, 2000, 1300, 1800, 1100, 1600, 900, 2200],
    'bedrooms': [2, 3, 2, 4, 3, 4, 2, 3, 1, 5],
    'price': [250000, 300000, 200000, 450000, 280000, 400000, 230000, 320000, 180000, 500000]
}
df = pd.DataFrame(data)
print(df.head())
```

### Split Dataset (70:30)

```python
train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)
print('Train shape:', train_df.shape)
print('Test shape:', test_df.shape)
```

### a. Normalize the required columns

```python
minmax_scaler = MinMaxScaler()
train_df[['area']] = minmax_scaler.fit_transform(train_df[['area']])
test_df[['area']] = minmax_scaler.transform(test_df[['area']])
print('--- After Normalization ---')
print(train_df.head())
```

### b. Standardize the required columns

```python
std_scaler = StandardScaler()
train_df[['bedrooms', 'price']] = std_scaler.fit_transform(train_df[['bedrooms', 'price']])
test_df[['bedrooms', 'price']] = std_scaler.transform(test_df[['bedrooms', 'price']])
print('--- After Standardization ---')
print(train_df.head())
```

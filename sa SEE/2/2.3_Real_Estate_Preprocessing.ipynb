```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('real_estate.csv')
df.head(10)
```

### Split Dataset (70:30)

```python
train_df, test_df = train_test_split(df, test_size=0.3, random_state=42)
print('Train shape:', train_df.shape)
print('Test shape:', test_df.shape)
```

### a. Normalize the required columns

```python
minmax_scaler = MinMaxScaler()
train_df[['area']] = minmax_scaler.fit_transform(train_df[['area']])
test_df[['area']] = minmax_scaler.transform(test_df[['area']])
train_df.head()
```

### b. Standardize the required columns

```python
std_scaler = StandardScaler()
train_df[['bedrooms', 'price']] = std_scaler.fit_transform(train_df[['bedrooms', 'price']])
test_df[['bedrooms', 'price']] = std_scaler.transform(test_df[['bedrooms', 'price']])
train_df.head()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
```

### a. Prepare Dataset (Mock Data)

```python
data = {
    'rainfall': [120, 150, 80, 200, 100],
    'temperature': [25, 28, 22, 30, 24],
    'Humidity': [70, 80, 60, 90, 65],
    'soil_moisture': [30, 40, 20, 50, 25]
}
df = pd.DataFrame(data)
print(df.head())
```

### a. Normalize the required columns

```python
minmax_scaler = MinMaxScaler()
df[['rainfall', 'temperature']] = minmax_scaler.fit_transform(df[['rainfall', 'temperature']])
print('--- After Normalization ---')
print(df.head())
```

### b. Standardize the required columns

```python
std_scaler = StandardScaler()
df[['Humidity', 'soil_moisture']] = std_scaler.fit_transform(df[['Humidity', 'soil_moisture']])
print('--- After Standardization ---')
print(df.head())
```

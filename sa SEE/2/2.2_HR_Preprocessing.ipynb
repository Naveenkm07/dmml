```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
```

### a. Prepare Dataset (Mock Data)

```python
data = {
    'salary': [50000, 60000, 45000, 70000, 55000, 80000, 52000, 65000, 48000, 72000],
    'years_of_experience': [2, 5, 1, 8, 3, 10, 2, 6, 1, 9]
}
df = pd.DataFrame(data)
print(df.head())
```

### Split Dataset (80:20)

```python
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
print('Train shape:', train_df.shape)
print('Test shape:', test_df.shape)
```

### a. Normalize the required columns

```python
minmax_scaler = MinMaxScaler()
train_df[['salary']] = minmax_scaler.fit_transform(train_df[['salary']])
test_df[['salary']] = minmax_scaler.transform(test_df[['salary']])
print('--- After Normalization ---')
print(train_df.head())
```

### b. Standardize the required columns

```python
std_scaler = StandardScaler()
train_df[['years_of_experience']] = std_scaler.fit_transform(train_df[['years_of_experience']])
test_df[['years_of_experience']] = std_scaler.transform(test_df[['years_of_experience']])
print('--- After Standardization ---')
print(train_df.head())
```

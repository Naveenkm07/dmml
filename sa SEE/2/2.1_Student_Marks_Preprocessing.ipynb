```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
```

### a. Prepare Dataset (Mock Data)

```python
data = {
    'Mathematics': [85, 90, 78, 92, 88],
    'Physics': [80, 85, 75, 89, 82],
    'Chemistry': [88, 92, 80, 95, 90]
}
df = pd.DataFrame(data)
print(df.head())
```

### a. Normalize the required columns

```python
minmax_scaler = MinMaxScaler()
df[['Mathematics', 'Physics']] = minmax_scaler.fit_transform(df[['Mathematics', 'Physics']])
print('--- After Normalization ---')
print(df.head())
```

### b. Standardize the required columns

```python
std_scaler = StandardScaler()
df[['Chemistry']] = std_scaler.fit_transform(df[['Chemistry']])
print('--- After Standardization ---')
print(df.head())
```

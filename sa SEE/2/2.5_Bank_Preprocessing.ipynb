```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
```

### a. Prepare Dataset (Mock Data)

```python
data = {
    'income': [60000, 40000, 80000, 30000, 90000, 50000, 70000, 45000, 85000, 35000],
    'loan_amount': [15000, 5000, 25000, 2000, 30000, 10000, 20000, 8000, 28000, 3000],
    'credit_score': [700, 600, 750, 550, 800, 650, 720, 620, 780, 580]
}
df = pd.DataFrame(data)
print(df.head())
```

### Split Dataset (75:25)

```python
train_df, test_df = train_test_split(df, test_size=0.25, random_state=42)
print('Train shape:', train_df.shape)
print('Test shape:', test_df.shape)
```

### a. Normalize the required columns

```python
minmax_scaler = MinMaxScaler()
train_df[['income']] = minmax_scaler.fit_transform(train_df[['income']])
test_df[['income']] = minmax_scaler.transform(test_df[['income']])
print('--- After Normalization ---')
print(train_df.head())
```

### b. Standardize the required columns

```python
std_scaler = StandardScaler()
train_df[['loan_amount', 'credit_score']] = std_scaler.fit_transform(train_df[['loan_amount', 'credit_score']])
test_df[['loan_amount', 'credit_score']] = std_scaler.transform(test_df[['loan_amount', 'credit_score']])
print('--- After Standardization ---')
print(train_df.head())
```

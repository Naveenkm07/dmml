```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
df = pd.read_csv('customer_purchase.csv')
print('BEFORE ENCODING:')
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

### a. Label Encoding on: ['Gender']

```python
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])
print('AFTER LABEL ENCODING:')
df.head()
```

### One-Hot Encoding on: ['Payment Method']

```python
onehot_encoder = OneHotEncoder()
encoded = onehot_encoder.fit_transform(df[['Payment Method']]).toarray()
df[onehot_encoder.get_feature_names_out(['Payment Method'])] = encoded
df.drop(columns=['Payment Method'], inplace=True)
print('AFTER ONE-HOT ENCODING:')
df.head()
```

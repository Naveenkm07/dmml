```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
df = pd.read_csv('employee_dataset.csv')
print('BEFORE ENCODING:')
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

### a. Label Encoding on: ['Department']

```python
label_encoder = LabelEncoder()
df['Department'] = label_encoder.fit_transform(df['Department'])
print('AFTER LABEL ENCODING:')
df.head()
```

### One-Hot Encoding on: ['Education Level', 'Job Role']

```python
onehot_encoder = OneHotEncoder()
encoded = onehot_encoder.fit_transform(df[['Education Level', 'Job Role']]).toarray()
df[onehot_encoder.get_feature_names_out(['Education Level', 'Job Role'])] = encoded
df.drop(columns=['Education Level', 'Job Role'], inplace=True)
print('AFTER ONE-HOT ENCODING:')
df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = pd.read_csv('car_evaluation.csv')
print('BEFORE ENCODING:')
df.head()
```

### One-Hot Encoding on: ['Buying Price', 'Maintenance Cost', 'Safety Rating']

```python
onehot_encoder = OneHotEncoder(sparse=False)
encoded = onehot_encoder.fit_transform(df[['Buying Price', 'Maintenance Cost', 'Safety Rating']])
df[onehot_encoder.get_feature_names_out(['Buying Price', 'Maintenance Cost', 'Safety Rating'])] = encoded
df.drop(columns=['Buying Price', 'Maintenance Cost', 'Safety Rating'], inplace=True)
print('AFTER ONE-HOT ENCODING:')
df.head()
```

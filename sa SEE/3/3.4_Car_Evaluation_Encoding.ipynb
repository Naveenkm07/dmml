```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
df = pd.read_csv('car_evaluation.csv')
print('BEFORE ENCODING:')
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

### Ordinal Encoding on: ['Buying Price', 'Maintenance Cost']

```python
ordinal_encoder = OrdinalEncoder(categories=[['Low', 'Medium', 'High'], ['Low', 'Medium', 'High']])
df[['Buying Price', 'Maintenance Cost']] = ordinal_encoder.fit_transform(df[['Buying Price', 'Maintenance Cost']])
print('AFTER ORDINAL ENCODING:')
df.head()
```

### One-Hot Encoding on: ['Safety Rating']

```python
onehot_encoder = OneHotEncoder()
encoded = onehot_encoder.fit_transform(df[['Safety Rating']]).toarray()
df[onehot_encoder.get_feature_names_out(['Safety Rating'])] = encoded
df.drop(columns=['Safety Rating'], inplace=True)
print('AFTER ONE-HOT ENCODING:')
df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = pd.read_csv('online_shopping.csv')
print('BEFORE ENCODING:')
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

### a. Label Encoding on: ['Visitor Type']

```python
label_encoder = LabelEncoder()
df['Visitor Type'] = label_encoder.fit_transform(df['Visitor Type'])
print('AFTER LABEL ENCODING:')
df.head()
```

### One-Hot Encoding on: ['Browser Type']

```python
onehot_encoder = OneHotEncoder(sparse=False)
encoded = onehot_encoder.fit_transform(df[['Browser Type']])
print('AFTER ONE-HOT ENCODING:')
print(encoded)
```

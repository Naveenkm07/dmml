```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder, OneHotEncoder
```

```python
df = pd.read_csv('student_performance.csv')
print('BEFORE ENCODING:')
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

### a. Label Encoding on: ['Grade Category']

```python
label_encoder = LabelEncoder()
df['Grade Category'] = label_encoder.fit_transform(df['Grade Category'])
print('AFTER LABEL ENCODING:')
df.head()
```

### Ordinal Encoding on: ['Gender', 'Study Level']

```python
ordinal_encoder = OrdinalEncoder()
df[['Gender', 'Study Level']] = ordinal_encoder.fit_transform(df[['Gender', 'Study Level']])
print('AFTER ORDINAL ENCODING:')
df.head()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = pd.read_csv('student_performance.csv')
print('BEFORE ENCODING:')
df.head()
```

```python
df.info()
```

```python
df.nunique()
```

### a. Label Encoding on: ['Grade Category']

```python
label_encoder = LabelEncoder()
df['Grade Category'] = label_encoder.fit_transform(df['Grade Category'])
print('AFTER LABEL ENCODING:')
df.head()
```

### One-Hot Encoding on: ['Gender', 'Study Level']

```python
onehot_encoder = OneHotEncoder(sparse=False)
encoded = onehot_encoder.fit_transform(df[['Gender', 'Study Level']])
df[onehot_encoder.get_feature_names_out(['Gender', 'Study Level'])] = encoded
df.drop(columns=['Gender', 'Study Level'], inplace=True)
print('AFTER ONE-HOT ENCODING:')
df.head()
```

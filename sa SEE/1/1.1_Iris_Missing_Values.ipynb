### a. Remove rows containing missing values

```python
df_dropped = df.dropna()
df_dropped.isnull().sum()
```

### b. Replace missing values with the mean of the column

```python
df_mean = df.fillna(df.mean(numeric_only=True))
df_mean.isnull().sum()
```

### c. Replace missing values with the median of the column

```python
df_median = df.fillna(df.median(numeric_only=True))
df_median.isnull().sum()
```

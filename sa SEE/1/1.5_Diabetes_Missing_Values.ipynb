### a. Remove rows containing missing values

```python
df_dropped = df.dropna()
df_dropped.isnull().sum()
```

### b. Replace missing values with the median of the column

```python
df_median = df.fillna(df.median(numeric_only=True))
df_median.isnull().sum()
```

### c. Replace missing values with the mode of the column

```python
df_mode = df.copy()
for col in df_mode.columns:
    df_mode[col].fillna(df_mode[col].mode()[0], inplace=True)
df_mode.isnull().sum()
```

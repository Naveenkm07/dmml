```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('ecommerce_fraud.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df['isFraud'].value_counts()
```

```python
x = df.drop('isFraud', axis=1)
y = df['isFraud']
y.value_counts()
```

### a. Random Oversampling

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=0)
x_resample, y_resample = ros.fit_resample(x, y)
print('After Random Oversampling:')
print(y_resample.value_counts())
```

### b. Random Undersampling

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=0)
x_resample1, y_resample1 = rus.fit_resample(x, y)
print('After Random Undersampling:')
print(y_resample1.value_counts())
```

### c. SMOTE

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
print('\nOriginal class distribution:', Counter(y))
smote = SMOTE(random_state=42)
X_resample2, y_resample2 = smote.fit_resample(x, y)
print('Resampled class distribution:', Counter(y_resample2))
```

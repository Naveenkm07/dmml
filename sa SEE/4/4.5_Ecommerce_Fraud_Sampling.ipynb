```python
import pandas as pd
import numpy as np
```

### Prepare Dataset (Mock Data)

```python
data = {
    'Device_Type': ['Mobile', 'Desktop', 'Mobile', 'Mobile', 'Desktop', 'Mobile'],
    'Payment_Method': ['Card', 'Card', 'PayPal', 'Card', 'Crypto', 'Crypto'],
    'Time_Spent': [10, 15, 12, 8, 2, 1],
    'isFraud': [0, 0, 0, 0, 1, 1]
}
df = pd.DataFrame(data)
print(df.head())
```

### Convert Categorical Variables

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Device_Type'] = le.fit_transform(df['Device_Type'])
df['Payment_Method'] = le.fit_transform(df['Payment_Method'])
print(df.head())
```

```python
x = df.drop('isFraud', axis=1)
y = df['isFraud']
print('--- Original Class Distribution ---')
print(y.value_counts())
```

### a. Random Oversampling

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=42)
x_ros, y_ros = ros.fit_resample(x, y)
print('--- After Random Oversampling ---')
print(y_ros.value_counts())
```

### b. Random Undersampling

```python
from imblearn.under_sampling import RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
x_rus, y_rus = rus.fit_resample(x, y)
print('--- After Random Undersampling ---')
print(y_rus.value_counts())
```

### c. SMOTE

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42, k_neighbors=1)
x_smote, y_smote = smote.fit_resample(x, y)
print('--- After SMOTE ---')
print(y_smote.value_counts())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
import sklearn.datasets

ds = sklearn.datasets.load_breast_cancer()
```

```python
ds.target_names
```

```python
df =pd.DataFrame(ds.data , columns=ds.feature_names)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
x= df
y=ds.target
```

```python
#to check the standard deviation of un standartized data , which will not be close to 1
print(ds.data.std())
```

```python
# standartize the data before splitting ,,, after standartize check the standard deviation fun, the result should be close to 1

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_standard = scaler.fit_transform(x)
print(x_standard)
print(x_standard.std())
```

```python
# Minmaxscaler transforms data to the range [0,1]
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
x_minmax=minmax.fit_transform(x)
print(x_minmax)
```
